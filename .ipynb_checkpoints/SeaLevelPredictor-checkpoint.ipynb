```python
import pandas as pd

df = pd.read_csv('Data/epa-sea-level.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np

year2050 = np.arange(171)+1880
res = linregress(df.iloc[:,:2])
idx = df[df['Year']==2000].index[0]
res2k = linregress(df.iloc[idx:,:2])

fig,ax = plt.subplots()
ax.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
ax.plot(year2050,res.intercept + res.slope*year2050, 'g', label='1880-2050 fit')
ax.plot(year2050[idx:],res2k.intercept + res2k.slope*year2050[idx:], 'r', label='2000-2050 fit')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
ax.legend()
```
