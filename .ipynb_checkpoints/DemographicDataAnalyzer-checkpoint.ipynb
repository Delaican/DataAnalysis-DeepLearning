```python
import pandas as pd
```

```python
colnames=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
df = pd.read_csv("Data/adult.data.csv", names=colnames, skipinitialspace = True)
df.head()
```

- How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels. (race column)

```python
race = df.pivot_table(index = ['race'], aggfunc ='size')
print(race,type(race))
```

- What is the average age of men?

```python
age_men = df.loc[df['sex'] == 'Male','age']
avg_age_men = sum(age_men)/len(age_men)
print(f'The average age of men is {int(avg_age_men)} years old.')
```

- What is the percentage of people who have a Bachelor's degree?

```python
bachelors_percentage = len(df.loc[df['education']=='Bachelors'])*100/len(df)
print(f'The percentage of people who have a Bachelor\'s degree is {round(bachelors_percentage,1)}%')
```

- What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
adv_edu = df.loc[(df['education']=='Bachelors') | (df['education']=='Masters') | (df['education']=='Doctorate')]
adv_edu_highincome = adv_edu.loc[adv_edu['income']=='>50K']
adv_edu_highincome_ptge = len(adv_edu_highincome)*100/len(adv_edu)
print(f'The percentage of people with advanced education (Bachelors, Masters, or Doctorate) that make more than 50K is {round(adv_edu_highincome_ptge,1)}%')
```

- What percentage of people without advanced education make more than 50K?

```python
notadv_edu = df.loc[(df['education']!='Bachelors') & (df['education']!='Masters') & (df['education']!='Doctorate')]
notadv_edu_highincome = notadv_edu.loc[notadv_edu['income']=='>50K']
notadv_edu_highincome_ptge = len(notadv_edu_highincome)*100/len(notadv_edu)
print(f'The percentage of people without advanced education that make more than 50K is {round(notadv_edu_highincome_ptge,1)}%')
```

- What is the minimum number of hours a person works per week?

```python
min_hours = min(df['hours-per-week'])
print(f'The minimum number of hours a person works per week is {min_hours}')
```

- What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
minhours = df.loc[df['hours-per-week']==min_hours]
minhours_highincome = minhours.loc[minhours['income']=='>50K']
minhours_highincome_ptge = len(minhours_highincome)*100/len(minhours)
print(f'The percentage of the people who work the minimum number of hours per week have a salary of more than 50K is {round(minhours_highincome_ptge,1)}%')
```

- What country has the highest percentage of people that earn >50K and what is that percentage?

```python
highincome = df.loc[df['income']=='>50K']
country = highincome.pivot_table(index = ['native-country'], aggfunc ='size')
high_country = country[country==max(country)]
high_country_ptge = high_country.values[0]*100/len(highincome)
print(f'The country with the highest percentage of people that earn >50K is {high_country.index[0]} with {round(high_country_ptge,1)}%')
```

- Identify the most popular occupation for those who earn >50K in India.

```python
india_high = highincome.loc[highincome['native-country']=='India']
india_high_ocu = india_high.pivot_table(index = ['occupation'], aggfunc ='size')
india_high_maxocu = india_high_ocu[india_high_ocu==max(india_high_ocu)]
print(f'the most popular occupation for those who earn >50K in India Is {india_high_maxocu.index[0]}')
```

## More analysis and Deep Learning Regression

Income column is a binary value (<=50K and >50K), therefore we can make it 0 and 1, respectively.

```python
df.loc[df['income']=='<=50K','income']=0
df.loc[df['income']=='>50K','income']=1
df['income'] = pd.to_numeric(df['income'])
```

A good initial insight of the dataset is knowing the basic statistical information about numerical features.

```python
df.describe()
```

Income has a mean of 0.24, meaning that approximately 25% of people in this dataset make >50K.

We can already search for correlations between numerical features, so we create a correlation matrix:

```python
import seaborn as sns
from matplotlib import pyplot as plt

data = df.select_dtypes(include='number')
#X = data.iloc[:,:-1]  #independent columns
#y = data.iloc[:,-1]    #target column i.e price range
#get correlations of each features in dataset
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

According to the correlation matrix, income is most correlated to education-num, which describes the level of education the individual has.

We can also search for patterns looking at the income with individual features. For instance, how many people gain more or less than 50K according to their age.

```python
plt.xlabel('age')
plt.ylabel('frequency')
plt.hist([df.loc[df['income']==1,'age'],df.loc[df['income']==0,'age']],label=['>50K','<=50K'],alpha=0.65)
plt.title('Income')
plt.legend()
```

```python
plt.xlabel('education')
plt.ylabel('frequency')
plt.hist([df.loc[df['income']==1,'education'],df.loc[df['income']==0,'education']],label=['>50K','<=50K'],alpha=0.65)
plt.xticks(rotation = 45)
plt.title('Income')
plt.legend()
```

We can see that more people win more than 50K in their 40s (approx 2000 people), however, there's still more people in their 40s to win less or equal to 50K (approx 4000).
